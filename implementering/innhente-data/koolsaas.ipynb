```python
import matplotlib.pyplot as plt
```

```python
fil = open("kolsaas.csv")
data = fil.read()
```

```python
linjer = data.split("\n")
tur = []
for linje in linjer:
    splittet_linje = linje.split(",")
    tur.append(splittet_linje)
tur[1][0]
```

```python
x = []
y = []
for linje in tur[1:]:
    høyde = linje[3]
    tid = linje[0]
    x.append(tid)
    y.append(høyde)
plt.plot(x,y)
plt.show()
```
